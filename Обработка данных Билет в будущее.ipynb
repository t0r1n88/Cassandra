```python
"""
Скрипт для нахождения классов для прошедших профпробы
Для этого нужно сравниьт 2 таблицы. по ключу ФИО, дело осложняется только тем что в одной таблице у меня есть только инициалы
4 студентов вообще без инициалов я вынес в отдельный файл.Можно будет найти отдельно. Также есть 6 фио со сложными отчествами
т.е. сплит по пробелу в них даст 4
"""
import pandas as pd



```

```python
def create_initials(cell):
    """
    Функция для создание инициалов из полного ФИО
    """
    lst_fio = cell.split()
    
    # Проверяем на длину получившегося списка
    if len(lst_fio) == 3:
        # Нормальный случай
        return f'{lst_fio[0]} {lst_fio[1][0]}. {lst_fio[2][0]}.'
    else:
        return 'Не удалось обработать'
    
def change_name_discipline(cell):
    # Функция для установки заглавной буквы для содержимого ячейки
    #очищаем от пробельных символов спереди и сзади
    value = cell.strip()
    # Делаем заглавным первый символ
    out_value = f'{value[0].upper()}{value[1:]}'

    return out_value
```

```python
# Обрабатывыем список  прошедших пробы


df_exe = pd.read_excel('Списки детей на профпробы полный.xlsx')

name_prob = ''

for row in df_exe.itertuples():
    # row[4] это название пробы
    if type(row[4]) == str :
        name_prob = row[4]
    df_exe.loc[row[0],'Наименование'] = name_prob
# Делаем первую букву заглавной
df_exe['Наименование'] = df_exe['Наименование'].apply(change_name_discipline)
```
